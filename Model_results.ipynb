# Visualization

```python
plt.figure(figsize=(10,5))
plt.title('Evolution of MAE with respect to the number of lookback days', size=20)
plt.plot([3,5,8,10,15,20,30],global_MAEs, color='k', label='LASSO baseline')
plt.plot([3,5,8,10,15,20,30],global_MAEs_, color='b', label='LASSO with mobility data')
plt.plot([3,5,8,10,15,20,30],global_MAEs_ridge, color='orange', label='RIDGE')
plt.xticks(size=15), plt.yticks(size=13)
plt.ylabel('MAE',size=20), plt.xlabel('Number of lookback days', size=15)
plt.legend()
plt.yscale('log')
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.title('Evolution of MAE with respect to the number of lookback days', size=20)
plt.plot([3,5,8,10,15,20,30],global_MAEs, color='k', label='LASSO baseline')
plt.plot([3,5,8,10,15,20,30],global_MAEs_, color='b', label='LASSO with mobility data')
plt.xticks(size=15), plt.yticks(size=13)
plt.ylabel('MAE',size=20), plt.xlabel('Number of lookback days', size=15)
plt.legend()
plt.show()
```

## Predictions

```python
preds_ridge = list(predictions_7d_ridge[predictions_7d_ridge.GeoID=='Germany'].PredictedDailyNewCases)
preds_ = list(predictions_7d_[predictions_7d_.GeoID=='Germany'].PredictedDailyNewCases)
preds_baseline = list(predictions_7d[predictions_7d.GeoID=='Germany'].PredictedDailyNewCases)
ground_truth = list(predictions_7d[predictions_7d.GeoID=='Germany'].NewConfirmedCases_7davg)

date = list(predictions_7d_ridge[predictions_7d_ridge.GeoID=='Germany'].Date)


plt.figure(figsize=(20,6))
plt.title('Predictions for Germany on 7 days', size=25)
plt.plot(date, preds_ridge, color='k', label='RIDGE')
plt.plot(date,preds_, color='b', label='LASSO with mobility data')
plt.plot(date,preds_baseline, color='orange', label='LASSO baseline')
plt.plot(date,ground_truth, label='GroundTruth', color='red',ls='-')
plt.xticks(size=16), plt.yticks(size=17)
plt.ylabel('New daily cases',size=20), plt.xlabel('Date', size=20)
plt.legend(loc=3,  prop={'size': 15})
#plt.yscale('log')
plt.show()
```
